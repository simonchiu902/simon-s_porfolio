The code is to crawl the vocabulary and its definition from the majortests website and store these words into a csv file.

```python
url = 'https://www.majortests.com/word-lists/word-list-0{0}.html'
def generate_urls(url, start_page, end_page):
    urls = [url.format(page) for page in range(start_page,end_page+1)]
    return urls
```

```python
import requests
def get_resource(url):
    headers = {"user-agent":"Mozilla/4.0 (compatible; MSIE 6.0; Windows NT 5.1; SV1; AcooBrowser; .NET CLR 1.1.4322; .NET CLR 2.0.50727)"}
    return requests.get(url, headers=headers)
```

```python
from bs4 import BeautifulSoup
def parse_html(html_str):
    return BeautifulSoup(html_str,'lxml')
```

```python
import time
def web_scraping_bot(urls):
    eng_words = []
    
    for url in urls:
        file = url.split('/')[-1]
        print('Crawlling '+ file + ' now')
        r = get_resource(url)
        if r.status_code == requests.codes.ok:
            soup = parse_html(r.text)
            words = get_words(soup, file)
            eng_words = eng_words + words
            print('waiting five seconds')
            time.sleep(5)
        else:
            print('HTTP error')
    return eng_words
```

```python
def get_words(soup, file):
    words = []
    count = 0
    for wordlist_table in soup.find_all(class_='wordlist'):
        count += 1
        for word_entry in wordlist_table.find_all('tr'):
            new_word = []
            new_word.append(file)
            new_word.append(count)
            new_word.append(word_entry.th.text)
            new_word.append(word_entry.td.text)
            words.append(new_word)
    return words
```

```python
def save_to_csv(words, file):
    import csv
    with open(file, 'w+', newline='', encoding='utf-8') as fp:
        writer = csv.writer(fp)
        for word in words:
            writer.writerow(word)
```

```python
if __name__ == '__main__':
    urls = generate_urls(url, 1, 3)
    eng_words = web_scraping_bot(urls)
    for item in eng_words:
        print(item)
    save_to_csv(eng_words,'majortests_words.csv')
```
